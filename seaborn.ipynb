```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
datos = sns.load_dataset("tips")
print(datos.head())
```

## grafico de pispercion, scatter plot

```python
sns.scatterplot(data=datos, x="total_bill", y="tip")
```

## histograma

```python
sns.histplot(data=datos, x="total_bill", bins=20)
```

## boxplot

```python
sns.boxplot(data=datos, x="day", y="total_bill")
```

## grafico de barras

```python
sns.barplot(data=datos, x="day", y="total_bill")
```

##  crear grafico de violin 

```python
sns.violinplot(data=datos, x="day", y="total_bill")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv('users.csv')

# Normalizar texto
df['pais'] = df['pais'].str.strip().str.lower()
df['genero'] = df['genero'].str.strip().str.lower()

# Filtrar usuarios de United Kingdom
uk_users = df[df['pais'] == 'united kingdom']

# Obtener los 20 usuarios ms viejos
top_20_uk = uk_users.sort_values(by='edad', ascending=False).head(20)

# Contar cuntos son male y female
conteo_genero = top_20_uk['genero'].value_counts()

# Crear grfico de barras
conteo_genero.plot(kind='bar', color=['lightcoral', 'skyblue'])

# Personalizar el grfico
plt.title('Gnero de los 20 usuarios ms viejos de United Kingdom')
plt.xlabel('Gnero')
plt.ylabel('Cantidad')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle='--', alpha=0.6)

# Mostrar grfico
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar el archivo CSV
df = pd.read_csv("users.csv")  # Asegrate de que el nombre sea correcto

# Filtrar por pases y edad
paises_deseados = ["Spain", "France", "Turkey"]
filtro = (df["pais"].isin(paises_deseados)) & (df["edad"] > 20)
usuarios_filtrados = df[filtro]

# Mostrar nombres de los usuarios filtrados
print("Nombres de usuarios mayores de 20 aos de Espaa, Francia o Turqua:")
print(usuarios_filtrados["nombre"])  # Cambia a "usuario" si esa es la columna

# Contar por gnero
conteo_genero = usuarios_filtrados["genero"].value_counts()

# Mostrar conteo
print("\nConteo por gnero:")
print(conteo_genero)

# Grfico
conteo_genero.plot(kind='bar', color=['lightblue', 'pink'])
plt.title("Usuarios > 20 aos de Espaa, Francia y Turqua por gnero")
plt.xlabel("Gnero")
plt.ylabel("Cantidad")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```
